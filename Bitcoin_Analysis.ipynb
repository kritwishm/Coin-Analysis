# Price change over time

```python
trace = go.Scatter(
    x=data.index,
    y=data['price(USD)'],
    name = "Price (USD)",
    line = dict(color = '#17BECF'),
    opacity = 0.8)

data1 = [trace]

layout = dict(
    title='Price(USD) v/s Time',
    xaxis=dict(
        title='Time',
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    ),
    yaxis=dict(
       title='Price(USD)'
    )
)

fig = dict(data=data1, layout=layout)
py.iplot(fig)
```

The Maximum Price attained by this coin is $19.47k on 16th Dec 2017. You can confirm it by moving the cursor on the graph.
